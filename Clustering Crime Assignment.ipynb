```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sn
```

```python
data = pd.read_csv("crime_data.csv")
data
```

```python
data.rename(columns={"Unnamed: 0": "Un"},inplace=True)
data.info()
```

```python
def norm_func(i):
    x = (i-i.min()) / (i.max() - i.min())
    return (x)
```

```python
df_norm = norm_func(data.iloc[:,1:])
df_norm
df =data.iloc[:,1:]
df = df.iloc[:,:-1]
df
```

```python
from sklearn.cluster import KMeans
fig = plt.figure(figsize=(10, 8))
WCSS = []
for i in range(1, 11):
    clf = KMeans(n_clusters=i)
    clf.fit(df_norm)
    WCSS.append(clf.inertia_)
plt.plot(range(1, 11), WCSS)
plt.title('The Elbow Method')
plt.ylabel('WCSS')
plt.xlabel('Number of Clusters')
plt.show()
```

```python
clf = KMeans(n_clusters=4)
y_kmeans = clf.fit_predict(df_norm)
```

```python
y_kmeans
clf.labels_
```

```python
md=pd.Series(y_kmeans)  
data['clust']=md  
data
```

```python
data.iloc[:,1:5].groupby(data.clust).mean()
```

```python
data.plot(x="Murder",y ="Assault",c=clf.labels_,kind="scatter",s=50 ,cmap=plt.cm.coolwarm) 
plt.title('Clusters using KMeans')
```

```python
data.plot(x="UrbanPop",y ="Rape",c=clf.labels_,kind="scatter",s=50 ,cmap=plt.cm.coolwarm) 
plt.title('Clusters using KMeans')
```

```python
from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch # for creating dendrogram 
#p = np.array(df_norm) # converting into numpy array format 
z = linkage(df_norm, method="complete",metric="euclidean")
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Murder')
sch.dendrogram(
    z,
    #leaf_rotation=0.,  # rotates the x axis labels
    #leaf_font_size=8.,  # font size for the x axis labels
)
plt.show()
```

### creating a  new column and assigning it to new column

```python
from sklearn.cluster import AgglomerativeClustering 
h_complete = AgglomerativeClustering(n_clusters=4, linkage='complete',affinity = "euclidean").fit(df_norm) 

cluster_labels=pd.Series(h_complete.labels_)
cluster_labels
data['clust']=cluster_labels 
data
```
